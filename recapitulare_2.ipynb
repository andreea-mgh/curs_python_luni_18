## Lista

```python
ce_am_mancat_azi = ["biscuite", "salata", "alune", "toast"]

# listele sunt ORDONATE (fiecare element are o pozitie)
print(ce_am_mancat_azi[0])

# listele sunt MUTABILE (putem adauga, sterge, edita elemente)
ce_am_mancat_azi[2] = "tofu"

# listele permit elemente duplicate
ce_am_mancat_azi.append("toast")
print(ce_am_mancat_azi)
```

## Set

```python
specii_de_animale = {'catelus','pisoi','labubu'}

# setul NU este ORDONAT
#print(specii_de_animale[0])

# se pot adauga/sterge elemente
specii_de_animale.add('iepure')
specii_de_animale.remove('pisoi')

# nu pot sa adaug un element de mai multe ori
specii_de_animale.add('catelus')
print(specii_de_animale)
```

# List ranges

```python
lista = [1, 2, 3, 5, 11, 14, 25, 27, 34]

# primele 4 elemente
print(lista[:4])

# toate elementele dupa elementul al doilea
print(lista[2:])

# de la elementul 3 pana la elementul 8 din 2 in 2
print(lista[2:8:2])

# toata lista din 2 in 2
print(lista[::2])

# lista invers
print(lista[::-1])

# ultimele 3 elemente
print(lista[-3:])
```

# List comprehensions

```python
lista = [1,3,5,6,12,8,4,13,22,11,9]
lista_pare = []
lista_patrate = []

for nr in lista:
    if nr % 2 == 0:
        lista_pare.append(nr)
    
for nr in lista:
    lista_patrate.append(nr*nr)

print(lista_pare)
print(lista_patrate)
```

```python
lista_pare = [nr for nr in lista if nr % 2 == 0]
print(lista_pare)

lista_patrate = [nr*nr for nr in lista]
print(lista_patrate)

lista_pare_la_patrat = [nr*nr for nr in lista if nr % 2 == 0]
print(lista_pare_la_patrat)
```
