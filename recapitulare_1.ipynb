## Exercitii

Scrie un program care afiseaza textul "mami imi cumperi robucsi" pana utilizatorul scrie "da".

```python
text = ""

while text != "da":
    print("mami imi cumperi robucsi?")
    text = input()
```

creeaza un program de "ghiceste numarul" (avem un numar secret si programul ne spune daca numarul ghicit e mai mare sau mai mic)

```python
nr_secret = 17

while True:
    nr = int(input())

    if nr == nr_secret:
        print("yayyyy bravo")
        break

    if nr > nr_secret:
        print("prea mare")
    
    else:
        print("prea mic")
```

suma numerelor de la 1 la n

```python
n = 15
s = 0

for x in range(1,n+1):
    s = s + x

print(s)
```

afiseaza un triunghi format din caracterul `*`

```python
for i in range(15):
    print('#' * (i+1))
```

Folosind `continue`, afiseaza toate numerele de la 1 la n, cu exceptia multiplilor lui 6.

!!! `continue` opreste bucla mai devreme, dar bucla inca continua de la inceput.

```python
n = 24

for i in range(1,n+1):
    if i%6 == 0:
        continue # face skip

    print(i)
```

cauta prima valoare negativa dintr-o lista

```python
lista = [12,34,1,0,-67,2,3,18,-10]

for nr in lista:
    if nr < 0:
        print(nr)
        break
```

numara vocalele dintr-un text

```python
text = "Ana are mere"
text = list(text.lower())
vocale = 0

for litera in text:
    if litera in "aeiou":
        vocale += 1

print(vocale)
 
```

### ELSE

La bucle, putem adauga o sectiune `else` iar codul din ea se va rula daca bucla s-a terminat fara intrerupere (`break`).

```python
incercari = 3
parola_buna = "biscuiti"

while incercari > 0:
    parola = input("Parola: ")

    if parola == parola_buna:
        print("Parola corecta!")
        break

    incercari -= 1
    print("Gresit!")
    
else:
    print("Cont blocat!")
```
