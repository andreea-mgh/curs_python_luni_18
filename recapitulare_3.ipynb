# Try-except

`ZeroDivisionError` cand impartim la 0

```python
try:
    numere = [6,4,3,0]
    for i in numere:
        print(12/i)
except ZeroDivisionError:
    print(":(")
```

`NameError` cand apelam o variabila care nu exista cu acel nume

```python
numar = 7
try:
    if Numar % 2 == 0:
        print("e par")
except Exception as e: # ca sa salvez textul erorii intr-o variabila (e)
    print("eroare: ", e)
```

`KeyError` cand caut o cheie in dictionar care nu exista

```python
preturi = {
    "suculet":4.30,
    "pufuleti":2.50,
    "alune":5.00
}

try:
    bani = 10
    produs = input("ce vrei sa cumperi? ")
    bani -= preturi[produs]
    print(f"ai luat {produs}, mai ai {bani} lei")
except KeyError:
    print("nu avem")
```

`ValueError` vs `TypeError`

```python
x = int('23')
print(x+2)

try:
    x = int('aaaaa')
except ValueError:
    print('string este un tip de date valid pentru functia int() dar valoarea nu e buna')

try:
    x = int([2,3])
except TypeError:
    print('list NU e un tip de date valid pentru functia int()')

try:
    x = 10 + '12'
except TypeError:
    print('tipul de date numar nu poate fi adunat cu tipul de date string (text)')

try:
    lista = [1,3,4,5]
    lista.remove(2)
except ValueError:
    print('functia remove are nevoie de numere, dar valoarea numarului e incorecta')
```

`FileNotFoundError`

```python
try:
    with open("fisier.txt", "r") as f:
        continut = f.read()
        print(continut)
except FileNotFoundError:
    print("fisierul nu exista")
```

blocurile `else` si `finally`

```python
e = 0

try:
    val = input("scrie un numar")
    val = int(val)
    x = 100 / val
except ValueError:
    e = 1
    print(f"{val} nu e un numar")
except ZeroDivisionError:
    e = 1
    print("nu pot imparti la zero")
else: # chestii care se ruleaza numai daca nu a dat eroare
    print(f"100 : {val} = {x}")
finally: # se ruleaza la final, indiferent ca a dat eroare sau nu
    print(f"codul s-a rulat cu {e} erori")
```
