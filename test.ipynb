Ce se va afișa după rularea următoarelor coduri?

```python
def fun(x):
    if x % 2 == 0:
        return 1
    else:
        return 2

print(fun(fun(2)))
```

```python
temp_list = [3, 1, -2]
# elementul cu indicele -1 este ULTIMUL
print(temp_list[ temp_list[-1] ])
```

```python
def functie(x):
    y = 1
    for i in range(1, x):
        if i % 3 == 0:
            y += i
    return y

print(functie(5))
```

```python
def func(x):
    return 3*x-7

print(func(func(func(4))))
```
